Question 8: Assign a 17th factor into the dataset, that is time of observation of cases (assign, an observation for each animal, if sampling occurs on Wednesdays, throughout the years. Remember there are 52 Wednesdays in a year).

```python
import datetime

# Generate random Wednesdays within 2023-2024
start_date = datetime.date(2023, 1, 1)
end_date = datetime.date(2024, 12, 31)

num_observations = len(df)
random_days = np.random.randint(0, (end_date - start_date).days, size=num_observations)

observation_dates = [start_date + datetime.timedelta(days=int(day)) for day in random_days]
wednesday_observations = [date for date in observation_dates if date.weekday() == 2]  # 2 represents Wednesday

# If you need exactly len(df) Wednesday observations:
while len(wednesday_observations) < num_observations:
    random_day = np.random.randint(0, (end_date - start_date).days)
    date = start_date + datetime.timedelta(days=random_day)
    if date.weekday() == 2:
        wednesday_observations.append(date)

# Assign to new column
df['Observation_Time'] = wednesday_observations[:num_observations]

print(df)
```

Question 12:
Develop another function which prints the data lines, sequentially in which SAARA is present, sorted based on whether a given ionophone antibiotic is administered or not. (This is an ANS 689 question. However, ANS 489 students are welcome to attend it).

```python
def print_data_by_ionophore(df):
    # Iterate through the dataset and check conditions
    for index, row in df.iterrows():
        rumen_modifier = row['Rumen_Modifier'].strip().lower()
        saara_present = row['Saara'].strip().lower() == 'yes'

        # Check ionophore and SAARA presence
        if (rumen_modifier == 'vm_mn' or rumen_modifier == 'mn' or rumen_modifier == 'vm') and saara_present:
            print(f"Observation {index + 1}:")
            print(row.to_string(), "\n")

# Call the function
print_data_by_ionophore(df)
```
